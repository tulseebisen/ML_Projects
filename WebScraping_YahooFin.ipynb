```python
import os
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings("ignore")
```

```python
url = "https://finance.yahoo.com/quote/000001.SS/history?p=000001.SS" #URl where data is stored

response = requests.get(url) # Pulling the Data to our environment 
```

```python
response.content # Lets see what Data we have
```

```python
soup = BeautifulSoup(response.content, "html.parser") # Sending Data to Cleaning Process.
```

```python
data = soup.find('table' , {'class' : 'W(100%) M(0)'}) #Finding req Data 
data = data.findAll('span') #tr,  span , 
# <tr styling> <span sflisafja> Data1 <\span> <tr>
# <tr styling> <span sflisafja> Data2 <\span> <tr>
```

```python
data
```

```python
dat = data[7: -3] # data sizing 
dat = [x.text for x in dat] # Taking only Text from Data
dat = np.array(dat).reshape(100,7)
```

```python
dat
```

```python
cols = data[:7]
cols = [col.text for col in cols]
cols
```

```python
df = pd.DataFrame(data = dat ,  columns = cols)
```
